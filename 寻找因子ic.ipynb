```python
#去极值，标准化处理
```

```python
def mad(factor):
    me = np.median(factor)
    mad = np.median(abs(factor-me))
    up = me + (3*1.4826*mad)
    down =me - (3*1.4826*mad)
    factor=np.where(factor>up,up,factor)
    factor=np.where(factor<down,down,factor)
    return factor

def stand(factor):
    mean=factor.mean()
    std=factor.std()
    return (factor-mean)/std
```

```python
#处理去极值，标准化
```

```python
mutiindex_df["pb_ratio"]=mad(mutiindex_df["pb_ratio"])
mutiindex_df["pb_ratio"]=stand(mutiindex_df["pb_ratio"])
```
